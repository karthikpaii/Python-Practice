```python
import requests
from bs4 import BeautifulSoup
```

```python
import requests

url = "https://www.savemylanguage.org/sml_v2/pages/dictionary.php"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept": "text/html"
}

response = requests.get(url, headers=headers, timeout=10)

soup = BeautifulSoup(response.text, "html.parser")
words = soup.find_all("span", class_="word-devanagari")
meanings = soup.find_all("span", class_="word-meaning")

print(f"{'Word Devanagari':<25} | Meaning")
print("-" * 60)

for w, m in zip(words, meanings):
    word = w.text.strip() or w.get("data-itrans")
    meaning = m.text.strip()
    print(f"{word:<25} | {meaning}")
```

```python
import csv

with open("dictionary.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)

    # Header
    writer.writerow(["Word Devanagari", "Meaning"])

    # Data
    for w, m in zip(words, meanings):
        word = w.text.strip() or w.get("data-itrans")
        meaning = m.text.strip()
        writer.writerow([word, meaning])

print("CSV file created successfully!")
```
